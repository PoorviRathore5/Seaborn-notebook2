```python
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.median,ci=None)
#np.mean,np.min,np.max these parameters can also be passed in the estimator
#to remove the errobar use ci=none
```

when there are multiple observations in each category,it also uses bootstrapping to compute a cconfidence interval around the estimate, which is plotted using error bars

```python
#pointplot 
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.median,ci=None)
```

```python
#countplot
sns.countplot(data=tips,x='sex',hue='day')
#hue is used to count for each day
```

its a special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable.This is similar to a histogram over a categorical,rather than quantitative,variable

```python
#faceting using catplot
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box')
```

```python
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

```python
#regplot -> axes level
#hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

In the simplest invocation,both functions(regplot,implot) draw a scatterplot of two variables,x and y, then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

## A second way to plot facet plots -> Facegrid

```python
# figure level -> relplot -> displot -> catplot -> implot

sns.catplot(data=tips, x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g= sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
g= sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
```

## Plotting Pairwise Relationship(PairGrid Vs Pairplot)

```python
iris
```

```python
sns.pairplot(iris,hue='species')   #it will automatically know which is neumerical and which is categorical column and will plot a graph accordingly
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
#g.map
g.map(sns.scatterplot)
```

```python
# map_diagonal -> map_offdiagonal

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)  #here we can plot violin plot/boxplot/scatterplot/2D histplot/2D kde plot/scatterplot/histplot
g.map_offdiag(sns.kdeplot) #here also
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
